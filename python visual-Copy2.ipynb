```python
plt.figure(figsize=(5,5))
plt.bar(plants,germinated,color='blue')
plt.title('Germination Rate')
plt.ylabel('No.of Seedings')
plt.xlabel('Plant names')
```

```python
plt.figure(figsize=(5,5))
plt.bar(plants,died,color='black')
plt.title('Failure Rate')
plt.ylabel('No.of Dead Seedings')
plt.xlabel('Plant names')
```

```python
plt.figure(figsize=(5,5))
plt.bar(plants,died,color='black',width=0.7)
plt.bar(plants,germinated,color='blue',align='edge',width=0.5)


plt.xlabel('Plant names')
```

```python
plt.figure(figsize=(5,5))
plt.bar(plants,died,color='black',bottom=germinated,label='Failed Seedlings')
plt.bar(plants,germinated,color='blue',label='Germinated Seedlings')
plt.legend()


plt.xlabel('Plant names')
plt.title("Greenhouse Seedlings")
```
