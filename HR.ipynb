**create testing and training of dataset and data cleaning**

```python
employee_df.head(3)
```

```python
#taking all the coulumns with non-numerical values to x_cat
x_cat=employee_df[['BusinessTravel','Department','EducationField','Gender','JobRole','MaritalStatus']]
x_cat
```

```python
#here onehotencoder transform the non-numerical data so that my AI can understand the data
from sklearn.preprocessing import OneHotEncoder
oneHotEncoder = OneHotEncoder()
x_cat=oneHotEncoder.fit_transform(x_cat).toarray()
```

```python
x_cat
```

```python
x_cat.shape
```

```python
x_cat=pd.DataFrame(x_cat)
```

```python
x_cat
```

```python
employee_df.columns
```

```python
#we dropped the target 'Atrittion', and all non numerical data('EducationField','Department','BusinessTravel','Gender','JobRole','MaritalStatus')
x_numerical=employee_df[['Age',  'DailyRate',
       'DistanceFromHome', 'Education',
       'EnvironmentSatisfaction',  'HourlyRate', 'JobInvolvement',
       'JobLevel',  'JobSatisfaction',
       'MonthlyIncome', 'MonthlyRate', 'NumCompaniesWorked', 'OverTime',
       'PercentSalaryHike', 'PerformanceRating', 'RelationshipSatisfaction',
       'StockOptionLevel', 'TotalWorkingYears', 'TrainingTimesLastYear',
       'WorkLifeBalance', 'YearsAtCompany', 'YearsInCurrentRole',
       'YearsSinceLastPromotion', 'YearsWithCurrManager']]
```

```python
x_all=pd.concat([x_cat,x_numerical],axis=1)
```

```python
#categorial data + numerical data
x_all
```

```python
x_all.columns = x_all.columns.astype(str)
```

```python
# Convert feature names to strings
x_all.columns = x_all.columns.astype(str)

# Scale the data
#here we are scaling the data because some features like DailyRate range are 1000s and other feature like DistanceFromHome,
#have values like less than 50 so without scaling DailyRate feature will take over other feature so we use MinMaxScaler
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x = scaler.fit_transform(x_all)

# Print the scaled data
print(x)
```

```python
x
```

```python
y=employee_df['Attrition']
```

**train and evaluate a LR Classifier**

```python
#here we are training 75% and testing 25% of dataset
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
y_pred =model.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix,classification_report
print('Accuracy {}%'.format(100 * accuracy_score(y_pred,y_test)))
```

```python
cm=confusion_matrix(y_pred,y_test)
sns.heatmap(cm,annot=True)
```

```python
print(classification_report(y_test,y_pred))
```

 train and evaluate the Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier

model=RandomForestClassifier()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
cm=confusion_matrix(y_pred,y_test)
sns.heatmap(cm,annot=True)
```

```python
print(classification_report(y_test,y_pred))
```

**train and evaluate a deep learning model**

```python
#relu-rectify linear units
import tensorflow as tf
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(units=500,activation='relu',input_shape=(50,)))
model.add(tf.keras.layers.Dense(units=500,activation='relu'))
model.add(tf.keras.layers.Dense(units=500,activation='relu'))
model.add(tf.keras.layers.Dense(units=1,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(optimizer ='Adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
epochs_hist=model.fit(x_train,y_train,epochs=100,batch_size=50)
```

```python
y_pred=model.predict(x_test)
y_pred=(y_pred>0.5)
y_pred
```

```python
plt.plot(epochs_hist.history['loss'])
plt.title('Model Loss Prog During Training')
plt.xlabel('Epoch')
plt.ylabel('Training Loss')
plt.legend(['Training Loss'])
```

```python
plt.plot(epochs_hist.history['accuracy'])
plt.title('Model accuracy Prog During Training')
plt.xlabel('Epoch')
plt.ylabel('Training accuracy')
plt.legend(['Training accuracy'])
```

```python
cm=confusion_matrix(y_test,y_pred)
sns.heatmap(cm,annot=True)
```

```python
print(classification_report(y_test,y_pred))
```
